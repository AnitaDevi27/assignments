```python
movie_data = pd.merge(ratings_data, movie_names, on='movieId')
movie_data.head()
```

```python
movie_data.groupby('title')['rating'].mean().head()
```

```python
#now sort by rating
movie_data.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
#count the ratings

movie_data.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
movie_data[['title', 'rating']]
```

```python
#calculate average ratings
ratings_mean_count = pd.DataFrame(movie_data.groupby('title')['rating'].mean())
```

```python
ratings_mean_count
```

```python
# add rating count
ratings_mean_count['rating_counts'] = pd.DataFrame(movie_data.groupby('title')['rating'].count())
ratings_mean_count.head()
```

```python
#round the rating
ratings_mean_count['rating'] = round(ratings_mean_count['rating'])
ratings_mean_count
```

```python
ratings_mean_count = ratings_mean_count[(ratings_mean_count['rating']>3) & (ratings_mean_count['rating_counts']>100)]
ratings_mean_count
```

```python
ratings_mean_count.count()
```

```python
# top 10 movies
ratings_mean_count.sort_values(by='rating', ascending = False).head(10)
```
